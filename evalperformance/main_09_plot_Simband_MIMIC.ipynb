```python
import os
import scipy.io as sio

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def my_plot_three_rows_ECG_PPG(mat_ECG_1, mat_PPG_1,
                               mat_ECG_2, mat_PPG_2,
                               mat_ECG_3, mat_PPG_3,
                               flip_int,
                               save_file_name):
    window = 30 # -sec

    fig = plt.figure(figsize=(7, 2))
    plt.tight_layout()
    plt.style.use(['seaborn-white', 'seaborn-paper'])
    lw = 1.0
    ax = fig.add_subplot(3, 1, 1)

    signal_ECG = mat_ECG_1['Ref_ECG_buffer']
    signal_ECG = (signal_ECG - min(signal_ECG)) / (max(signal_ECG) - min(signal_ECG))
    x_t_ECG = np.linspace(0, window , signal_ECG.shape[0])
    ax.plot(x_t_ECG, signal_ECG, color='#9c9c9c', ls='solid',linewidth=lw)

    signal_PPG = mat_PPG_1['PPG_buffer'] * flip_int[0]
    signal_PPG = (signal_PPG - min(signal_PPG)) / (max(signal_PPG) - min(signal_PPG))
    x_t_PPG = np.linspace(0, window , signal_PPG.shape[0])
    ax.plot(x_t_PPG, signal_PPG, color='k', ls='solid',linewidth=lw)

    ax.set_xlabel('',fontsize=30)
    ax.set_ylabel('',fontsize=30)
    ax.tick_params(labelsize=20)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])

    ax = fig.add_subplot(3, 1, 2)

    signal_ECG = mat_ECG_2['Ref_ECG_buffer']
    signal_ECG = (signal_ECG - min(signal_ECG)) / (max(signal_ECG) - min(signal_ECG))
    x_t_ECG = np.linspace(0, window , signal_ECG.shape[0])
    ax.plot(x_t_ECG, signal_ECG, color='#9c9c9c', ls='solid',linewidth=lw)

    signal_PPG = mat_PPG_2['PPG_buffer'] * flip_int[1]
    signal_PPG = (signal_PPG - min(signal_PPG)) / (max(signal_PPG) - min(signal_PPG))
    x_t_PPG = np.linspace(0, window , signal_PPG.shape[0])
    ax.plot(x_t_PPG, signal_PPG, color='k', ls='solid',linewidth=lw)

    ax.set_xlabel('',fontsize=30)
    ax.set_ylabel('',fontsize=30)
    ax.tick_params(labelsize=20)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])

    ax = fig.add_subplot(3, 1, 3)

    signal_ECG = mat_ECG_3['Ref_ECG_buffer']
    signal_ECG = (signal_ECG - min(signal_ECG)) / (max(signal_ECG) - min(signal_ECG))
    x_t_ECG = np.linspace(0, window , signal_ECG.shape[0])
    ax.plot(x_t_ECG, signal_ECG, color='#9c9c9c', ls='solid',linewidth=lw)

    signal_PPG = mat_PPG_3['PPG_buffer'] * flip_int[2]
    signal_PPG = (signal_PPG - min(signal_PPG)) / (max(signal_PPG) - min(signal_PPG))
    x_t_PPG = np.linspace(0, window , signal_PPG.shape[0])
    ax.plot(x_t_PPG, signal_PPG, color='k', ls='solid',linewidth=lw)

    ax.set_xlabel('',fontsize=30)
    ax.set_ylabel('',fontsize=30)
    ax.tick_params(labelsize=20)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])

    path_plot = r'/mnt/r/ENGR_Chon/Dong/Postdoc/ppg_open_data/my_paper/NPJ_digital_medicine/Figures'
    plt.savefig(os.path.join(path_plot,save_file_name+'.png'), dpi=300, bbox_inches='tight',pad_inches = 0.1)
```
