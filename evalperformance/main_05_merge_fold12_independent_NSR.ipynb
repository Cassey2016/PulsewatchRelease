```python
"""
Merge the fold-1,2 and independent NSR results together based on number of subjects.
Dong, 09/13/2024.
"""
import os
import pandas as pd
path_fold12_test = r'/mnt/r/ENGR_Chon/Dong/Python_generated_results/deep_learning_2023/analysis_2024_07_30/fold_1_2_combined_subject_independent_test'
filename_fold12_test = 'fold_1_2_combined_subject_independent_test_all_models.csv'

path_independentNSR_test = r'/mnt/r/ENGR_Chon/Dong/Python_generated_results/deep_learning_2023/analysis_2024_07_30/remain_NSR_combined_subject_independent_test'
filename_independentNSR_test = 'remain_NSR_combined_subject_independent_test_all_models.csv'

df_fold12 = pd.read_csv(os.path.join(path_fold12_test,filename_fold12_test))
df_indpNSR = pd.read_csv(os.path.join(path_independentNSR_test,filename_independentNSR_test))

num_fold12_sub = 72
num_indpNSR_sub = 34
```

```python
# Only consider the TPR and TNR
def my_avg_fold12_indpNSR(df_fold12,df_indpNSR,target_class):
    if target_class == 'NSR':
        col_name = 'TPR'
        class_num = 0
    elif target_class == 'AF':
        col_name = 'TNR'
        class_num = 1
    elif target_class == 'PACPVC':
        col_name = 'TNR'
        class_num = 2

    unique_model_idx = df_fold12['str_model_idx'].unique()
    for this_model_idx in unique_model_idx:
        temp_df_fold12_NSR_idx = df_fold12.index[(df_fold12['str_model_idx'] == this_model_idx) & (df_fold12['class'] == class_num)].tolist()
        # print(temp_df_fold12_NSR_idx)

        temp_df_indpNSR_NSR_idx = df_indpNSR.index[(df_indpNSR['str_model_idx'] == this_model_idx) & (df_indpNSR['class'] == class_num)].tolist()
        # print(temp_df_indpNSR_NSR_idx)

        # print(sum([df_fold12.at[temp_df_fold12_NSR_idx[0],'TPR'], df_indpNSR.at[temp_df_indpNSR_NSR_idx[0],'TPR']])/2)
        df_fold12.at[temp_df_fold12_NSR_idx[0],col_name] = sum([df_fold12.at[temp_df_fold12_NSR_idx[0],col_name], df_indpNSR.at[temp_df_indpNSR_NSR_idx[0],col_name]])/2
        # break

    return df_fold12

df_fold12 = my_avg_fold12_indpNSR(df_fold12,df_indpNSR,target_class='NSR')
df_fold12 = my_avg_fold12_indpNSR(df_fold12,df_indpNSR,target_class='AF')
df_fold12 = my_avg_fold12_indpNSR(df_fold12,df_indpNSR,target_class='PACPVC')

path_output = r'/mnt/r/ENGR_Chon/Dong/Python_generated_results/deep_learning_2023/analysis_2024_07_30/merge_fold12_indpNSR'
df_fold12.to_csv(os.path.join(path_output, 'merge_fold12_indpNSR_independent_test_all_models.csv'))
```
