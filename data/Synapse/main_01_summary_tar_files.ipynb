```python
"""
Summarizing the tar file names, tar file sizes, and last modified date in each folder under Synapse root folder.
Output is the "tar_files_XX_XXX.csv" file. 
Dong, 02/10/2025.
"""

def my_get_file_details(path_tar_files,path_output,filename_output):
    from pathlib import Path
    from datetime import datetime
    import hashlib

    #...Your code getting path here...

    directory = Path(path_tar_files)
    paths = []
    filename = []
    size = []
    hashes = []
    modified = []
    files = list(directory.glob('**/*.*'))

    for file in files:
        paths.append(file.parents[0])
        filename.append(file.parts[-1])
        size.append(file.stat().st_size)
        modified.append(datetime.fromtimestamp(file.stat().st_mtime))

    import pandas as pd
    df_file_details = pd.DataFrame(
        {'filename': filename,
        'filesize': size,
        'lastmodified': modified
        })

    import os
    Path(path_output).mkdir(parents=True, exist_ok=True)
    df_file_details.to_csv(os.path.join(path_output,filename_output), index=False)
```
